```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Customer_Car_Purchase.csv')

x = data.iloc[:,:-1].values
y = data.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.svm import SVC

classifier_linear = SVC(kernel='linear',random_state=0)
classifier_linear.fit(x_train,y_train)

classifier_rbf = SVC(kernel='rbf',random_state=0)
classifier_rbf.fit(x_train,y_train)

classifier_sigmoid = SVC(kernel='sigmoid',random_state=0)
classifier_sigmoid.fit(x_train,y_train)

classifier_poly = SVC(kernel='poly',random_state=0)
classifier_poly.fit(x_train,y_train)


#The portion below is for Support Vector Regression
from sklearn.svm import SVR
regressor = SVR(kernel='rbf')
regressor.fit(x_train,y_train)
```

```python
y_pred_linear = classifier_linear.predict(x_test)
y_pred_rbf = classifier_rbf.predict(x_test)
y_pred_sigmoid = classifier_sigmoid.predict(x_test)
y_pred_poly = classifier_poly.predict(x_test)

#The snippet below is for prediction based on Support Vector Regression
y_pred_svr = regressor.predict(x_test)
a =[]
for i in range(len(y_pred_svr)):
    a.append([y_test[i],y_pred_svr[i]])

for i in a:
    print(i)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix

print("Metrics for Linear:")
print("Accuracy: ",accuracy_score(y_test,y_pred_linear))
print("Confusion Matrix:")
print(confusion_matrix(y_test,y_pred_linear))
print('\n')
print("Metrics for RBF:")
print("Accuracy: ",accuracy_score(y_test,y_pred_rbf))
print("Confusion Matrix:")
print(confusion_matrix(y_test,y_pred_rbf))
print('\n')
print("Metrics for Poly:")
print("Accuracy: ",accuracy_score(y_test,y_pred_poly))
print("Confusion Matrix:")
print(confusion_matrix(y_test,y_pred_poly))
print('\n')
print("Metrics for Sigmoid:")
print("Accuracy: ",accuracy_score(y_test,y_pred_sigmoid))
print("Confusion Matrix:")
print(confusion_matrix(y_test,y_pred_sigmoid))
# print('\n')
# print("Metrics for Support Vector Regression:")
# # print("Accuracy: ",accuracy_score(y_test,y_pred_svr))
# print("Confusion Matrix:")
# print(confusion_matrix(y_test,y_pred_svr))
```
