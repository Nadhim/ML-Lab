```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/data.csv.txt')
```

#1. Line Plot


```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='Date', y='Value1', data=df, marker='o')
plt.title('Line Plot')
plt.xlabel('Date')
plt.ylabel('Value1')
plt.xticks(rotation=45)
plt.show()
```

#2. Bar Plot

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Value1', data=df)
plt.title('Bar Plot')
plt.xlabel('Category')
plt.ylabel('Value1')
plt.show()
```

#3. Histogram

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Value1'], bins=5, kde=True)
plt.title('Histogram')
plt.xlabel('Value1')
plt.ylabel('Frequency')
plt.show()
```

#4. Scatter Plot

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Value1', y='Value2', data=df, hue='Category')
plt.title('Scatter Plot')
plt.xlabel('Value1')
plt.ylabel('Value2')
plt.show()
```

# 5. Box Plot

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Category', y='Value1', data=df)
plt.title('Box Plot')
plt.xlabel('Category')
plt.ylabel('Value1')
plt.show()
```

#6. Pie Chart

```python
plt.figure(figsize=(8, 8))
df['Category'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Pie Chart')
plt.ylabel('')
plt.show()
```

#7. Heat Map

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Heatmap')
plt.show()
```

#8. Area Chart

```python
plt.figure(figsize=(10, 6))
df.groupby('Date')['Value1'].sum().plot(kind='area', stacked=False)
plt.title('Area Chart')
plt.xlabel('Date')
plt.ylabel('Value1')
plt.show()
```

# 9. Violin plot

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='Category', y='Value1', data=df)
plt.title('Violin Plot')
plt.xlabel('Category')
plt.ylabel('Value1')
plt.show()
```

# 10. Bubble plot

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Value1', y='Value2', size='Value2', data=df)
plt.title('Bubble Plot')
plt.xlabel('Value1')
plt.ylabel('Value2')
plt.show()
```
